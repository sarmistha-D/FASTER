```python

import os

train_dir = os.listdir('dataset/train')
test_dir = os.listdir('dataset/test')

train_frames = {}

for val in train_dir:
    frames = os.listdir('dataset/train/' + val)
    frames_list = ['dataset/train/' + val + '/' + seg for seg in frames if seg.endswith('.jpg')]
    if len(frames_list) > 0:
        train_frames[val] = frames_list

test_frames = {}

# Filter for .jpg files in the test dataset
for val in test_dir:
    frames = os.listdir('dataset/test/' + val)
    frames_list = ['dataset/test/' + val + '/' + seg for seg in frames if seg.endswith('.jpg')]
    if len(frames_list) > 0:
        test_frames[val] = frames_list
```

```python

import os

label_train_dir = os.listdir('dataset_2/train')
label_test_dir = os.listdir('dataset_2/test')

label_train_frames = {}
for val in train_dir:
    frames = os.listdir('dataset_2/train/' + val)
    frames_list = []

    for seg in frames:
        if seg.endswith('.jpg'):  # Only consider .jpg files
            try:
                # Append the index of the corresponding frame from train_frames
                frames_list.append(train_frames[val].index('dataset/train/' + val + '/' + seg))
            except:
                # Skip frames that are not found
                pass

    if len(frames_list) > 0:
        label_train_frames[val] = frames_list

label_test_frames = {}
for val in test_dir:
    frames = os.listdir('dataset_2/test/' + val)
    frames_list = []

    for seg in frames:
        if seg.endswith('.jpg'):  # Only consider .jpg files
            try:
                # Append the index of the corresponding frame from test_frames
                frames_list.append(test_frames[val].index('dataset/test/' + val + '/' + seg))
            except:
                # Skip frames that are not found
                pass

    if len(frames_list) > 0:
        label_test_frames[val] = frames_list

```
